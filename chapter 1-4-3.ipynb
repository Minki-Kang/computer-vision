1.4.3 Geometric Transformations of Images

Scaling

```python
import cv2
import numpy as np
img = cv2.imread('sonny.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
```

```python
cv2.imshow('image',img)
cv2.imshow('res image',res)
cv2.waitKey(0)
cv2.destroyAllWindows()
height, width = img.shape[:2]
net = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_NEAREST)
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)
cv2.imshow('image',img)
cv2.imshow('res image',res)
cv2.imshow('net image',net)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Translation

```python
import cv2
import numpy as np
img = cv2.imread('mario.jpg',1)
rows,cols,ch = img.shape
tx=100
ty=100
M = np.float32([[1,0,tx],[0,1,ty]])
dst = cv2.warpAffine(img,M,(cols,rows))
dst =cv2.rectangle(dst,(0,0),(tx,ty),(0,255,0),-1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Rotation

```python
img = cv2.imread('sonny.jpg',1)
rows,cols,ch = img.shape
cx=100
cy=100
thaet=50
M = cv2.getRotationMatrix2D((cx,cy),thaet,1)
dst = cv2.warpAffine(img,M,(cols,rows))
dst =cv2.circle(dst,(cx,cy),10,(0,255,0),-1)
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Affine Transformation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('dsu4.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

Perspective Transformation

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('dsu5.jpg',1)
rows,cols,ch= img.shape
pts1 = np.float32([[579,336],[754,376],[762,504],[583,488]])
pts2 = np.float32([[570,336],[754,335],[762,496],[570,488]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(0,0))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
